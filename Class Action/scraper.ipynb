```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import time
import pandas as pd
import time

# Specify the path to ChromeDriver (Ensure this path is correct)
chrome_driver_path = "C:\\WebDrivers\\chromedriver.exe"  # Update this path

# Set Chrome options to attach to the existing session
chrome_options = Options()
chrome_options.debugger_address = "127.0.0.1:9222"  # Connect to existing Chrome session

# Initialize Selenium WebDriver with the existing Chrome session
service = Service(chrome_driver_path)
driver = webdriver.Chrome(service=service, options=chrome_options)

# Define base URL (modify based on website's pagination structure)
base_url = "https://securities.stanford.edu/filings.html?page="  # Update with correct website

# List to store all extracted table data
all_data = []

# Loop through pages 1 to 3
for page in range(1, 55):
    url = f"{base_url}{page}"
    print(f"Accessing: {url}")
    driver.get(url)  # Navigate to the webpage
    time.sleep(2)  # Allow page to load

    # Find the table element (Modify this selector if needed)
    try:
        table = driver.find_element(By.TAG_NAME, "table")
        rows = table.find_elements(By.TAG_NAME, "tr")

        # Extract table headers
        headers = [th.text.strip() for th in rows[0].find_elements(By.TAG_NAME, "th")]

        # Extract table rows
        for row in rows[1:]:  # Skip header row
            columns = row.find_elements(By.TAG_NAME, "td")
            data = [col.text.strip() for col in columns]
            all_data.append(data)

    except Exception as e:
        print(f"Error extracting table from page {page}: {e}")

# Convert data to DataFrame
df = pd.DataFrame(all_data, columns=headers)

# Save to CSV
csv_filename = "extracted_table_data.csv"
df.to_csv(csv_filename, index=False, encoding="utf-8")
print(f"Table data saved to {csv_filename}")

# Close browser session
driver.quit()
```
